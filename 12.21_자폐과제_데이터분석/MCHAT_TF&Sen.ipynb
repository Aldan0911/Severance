# TfidVectorizer

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import linear_kernel
```

```python
# 1. Load the data from the original Excel file
# Load the data from the provided file again
cdb_csv = pd.read_csv(r"C:\Users\user\Desktop\vsc\Severance_Hospital\12.21_자폐과제_데이터분석\original\cdb_advanced_ASD_F3.csv")


# 2. Text Preprocessing
def preprocess_text(text):
    return " ".join([word.lower() for word in text.split()])
cdb_csv['Processed_Name'] = cdb_csv['Name'].apply(preprocess_text)
```

```python
# 3. Vectorization using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X_full = vectorizer.fit_transform(cdb_csv['Processed_Name'])


# 4. KMeans clustering with 100 clusters (assuming it was done earlier)
kmeans_100 = KMeans(n_clusters=100, random_state=42).fit(X_full)
cdb_csv['100_Cluster'] = kmeans_100.predict(X_full)
```

```python
# Function to find best matches in dataset based on the processed text and include rank : TFIDF
def match_question_to_data_detailed(question, cdb_csv, vectorizer, top_n=None):
    question_vec = vectorizer.transform([preprocess_text(question)])
    cosine_similarities = linear_kernel(question_vec, X_full).flatten()
    if top_n:
        related_docs_indices = cosine_similarities.argsort()[-top_n:][::-1]
    else:
        related_docs_indices = cosine_similarities.argsort()[::-1]
    matched_data = cdb_csv.iloc[related_docs_indices]
    matched_data['Question']= question
    matched_data['Matching Rank/Probability'] = cosine_similarities[related_docs_indices]
    return matched_data
# 6. List of Translated Questions
# questions_english = [
#     "When you call your child's name, does the child look at you?",
#     "How easy is it for you to make eye contact with your child?",
#     "When your child plays alone, do they line up objects?",
#     "Do other people easily understand what your child says?",
#     "Does your child point with their finger to express what they want (e.g., a toy that's far away)?",
#     "Does your child point with their finger to share interest with you (e.g., at an interesting sight)?"
# ] # df_mchat_eg_F['Text']

#questions_english = df_mchat_eg_F['Text']
```
