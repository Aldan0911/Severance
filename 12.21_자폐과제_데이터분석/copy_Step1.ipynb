```python
# Open the PDF file in binary mode
with open(mchat_en_path, 'rb') as pdf_file:
    # Create a PDF object
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Initialize an empty list to store the text lines
    text_lines = []

    # Loop through each page in the PDF
    for page in pdf_reader.pages:
        # Extract text from the page and split it into lines
        lines = page.extract_text().split('\n')
        
        # Append each line to the text_lines list
        text_lines.extend(lines)

# Create a DataFrame from the text_lines list
df_mchat_eg = pd.DataFrame({'Text': text_lines})
# Display the DataFrame
df_mchat_eg
```

# M-CHAT_kr

```python
mchat_kr_pt = r"C:\Users\user\Desktop\vsc\Severance_Hospital\12.21_자폐과제_데이터분석\M-CHAT_kr.pdf"
# Open the PDF file in binary mode
with open(mchat_kr_pt , 'rb') as pdf_file:
    # Create a PDF object
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    # Initialize an empty string to store the text
    text = ''

    # Loop through each page in the PDF
    for page_num in range(len(pdf_reader.pages)):
        # Extract text from the page and add a newline character between pages
        page = pdf_reader.pages[page_num]
        text += page.extract_text() + '\n'

# Use regular expression to split the text based on numbers
text_lines = re.split(r'\d+\s*', text.strip())

# Create a DataFrame from the text_lines list
mchat_kr = pd.DataFrame({'Text': text_lines})
mchat_kr
```
