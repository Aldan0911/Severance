```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.preprocessing import MinMaxScaler
```

```python
# seperating datasets
target_column = "Mutation"
numerical_column = train.columns.drop(target_column)
output_rows = train[target_column]
train.drop(target_column, axis=1, inplace=True)
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
t_train = scaler.transform(train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(t_train, output_rows, test_size=0.25, random_state=0)
```

```python
print('X_train:',np.shape(X_train))
print('y_train:',np.shape(y_train))
print('X_test:',np.shape(X_test))
print('y_test:',np.shape(y_test))
```

```python
basic_model = Sequential()
```

```python
basic_model.add(Dense(units=16, activation='relu', input_shape=(13,)))
```

```python
basic_model.add(Dense(1, activation='sigmoid'))
```

```python
adam = keras.optimizers.Adam(learning_rate=0.001)
```

```python
basic_model.compile(loss='binary_crossentropy', optimizer=adam, metrics=["accuracy"])
```

```python
basic_model.fit(X_train, y_train, epochs=100)
```
